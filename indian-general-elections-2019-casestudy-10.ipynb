**This project focuses on conducting Exploratory Data Analysis (EDA) using any analytical tool for the 2019 Indian General Elections dataset. The dataset, available on Kaggle and created by Prakrut Chauhan, serves as the foundation for the analysis.
The dataset encompasses comprehensive details about candidates who participated in the elections across different constituencies. It encompasses both personal attributes like assets, education, and criminal records, and electoral attributes like the constituency they contested from their affiliated political party, the total count of received votes, and more.
Throughout the project, we leverage a variety of analytical tools to cleanse the data and create insightful visualizations. The objective is to uncover patterns, trends, and relationships within the dataset to extract meaningful insights about the 2019 Indian General Elections.**

```python
#let impoerting a libraries as
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Exploratory Analysis and Visualization**

# datatype corrections

```python
df['CRIMINAL\nCASES'].fillna(0, inplace=True)
```

```python
df['CRIMINAL\nCASES'].unique()
```

```python
# Replace 'Not Available' with 0 in 'CRIMINAL\nCASES'
df['CRIMINAL\nCASES'].replace('Not Available', 0, inplace=True)
```

```python
# Convert 'CRIMINAL\nCASES' to integer
df['CRIMINAL\nCASES'] = df['CRIMINAL\nCASES'].astype(int)
```

```python
df['CRIMINAL\nCASES'].dtype
```

```python
#TO CHECK DATATYPE USING
df["AGE"].unique()
```

```python
# Replace NaN values with mean values in 'AGE'
mean_age = df['AGE'].mean()
```

```python
df['AGE'] = df['AGE'].fillna(mean_age)
```

```python
# Convert 'AGE' to integer
df["AGE"]=df["AGE"].astype(int)
```

```python
#TO CHECK DATATYPE USING
df["AGE"].dtype
```

# REGION MAPPING

```python
india = {
    'NORTH INDIA': ['Jammu & Kashmir', 'Himachal Pradesh', 'Punjab', 'Haryana', 'Uttarakhand', 'NCT OF Delhi', 'Chandigarh'],
    'CENTRAL INDIA': ['Uttar Pradesh', 'Madhya Pradesh', 'Chhattisgarh'],
    'WEST INDIA': ["Gujarat", "Rajasthan", "Maharashtra", "Goa", "Dadra & Nagar Haveli", "Daman & Diu"],
    'SOUTH INDIA': ['Karnataka', 'Andhra Pradesh', 'Telangana', 'Kerala', 'Tamil Nadu', 'Puducherry', 'Lakshadweep'],
    'EAST INDIA': ["Bihar", "Jharkhand", "West Bengal", "Odisha", "Andaman & Nicobar Islands"], 
    'NORTH-EAST INDIA': ['Assam', 'Arunachal Pradesh', 'Manipur', 'Meghalaya', 'Mizoram', 'Nagaland', 'Sikkim', 'Tripura']
}
```

```python
# Map states to regions
df['REGION'] = df['STATE'].apply(lambda state: next(region for region, states in india.items() if state in states))
```

```python
df.columns
```

# Seat Share

```python
#PLOTTING THE CHART 
plt.figure(figsize=(12, 10))
sns.countplot(x='REGION', hue='PARTY', data=df[df['WINNER'] == 1])
plt.title('Seat Share by Region and Party')
plt.xlabel('Region')
plt.ylabel('Number of Seats')
plt.legend()
plt.show()
```

# Age

```python
#plotting the histgram chart for AGE
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='AGE', bins=20, kde=True,color="salmon")
plt.title('Age Distribution of Elected MPs')
plt.xlabel('Age')
plt.ylabel('Number of MPs')
plt.show()
```

# Seat Category

# 1.Which States/UTs and Constituencies had the highest and the lowest Voter Turnout?

```python
# Voter Turnout by States/UTs and Constituencies
highest_turnout_states = df.groupby('STATE')['OVER TOTAL VOTES POLLED \nIN CONSTITUENCY'].mean().sort_values(ascending=False).head()
```

```python
lowest_turnout_states = df.groupby('STATE')['OVER TOTAL VOTES POLLED \nIN CONSTITUENCY'].mean().sort_values().head()
```

```python
print(highest_turnout_states)
```

```python
print(lowest_turnout_states)
```

```python
# Plotting Highest Turnout
plt.figure(figsize=(10, 6))
highest_turnout_states.plot(kind="bar")
plt.title('State-wise Highest Voter Turnout')
plt.xlabel('State')
plt.ylabel('Voter Turnout')
plt.xticks(rotation=0)
plt.show()

# Plotting Lowest Turnout
plt.figure(figsize=(10, 6))
lowest_turnout_states.plot(kind="bar")
plt.title('State-wise Lowest Voter Turnout')
plt.xlabel('State')
plt.ylabel('Voter Turnout')
plt.xticks(rotation=0)
plt.show()
```

# 2.How many elected MPs have a criminal record and what is their party-wise distribution?

```python
# Count the number of elected MPs with a criminal record
criminal_counts = df[df['CRIMINAL\nCASES'] > 0].groupby('PARTY')['CRIMINAL\nCASES'].count().reset_index()
```

```python
top10_party=criminal_counts.sort_values(by='CRIMINAL\nCASES',ascending=False).head(10)
```

```python
# Plotting top 10 party-wise distribution of elected MPs with a criminal record
plt.figure(figsize=(15, 6))
sns.barplot(x='PARTY', y='CRIMINAL\nCASES', data=top10_party, palette='rocket')
plt.title('Party-wise Distribution of Elected MPs with Criminal Record')
plt.xlabel('Party')
plt.ylabel('Number of MPs with Criminal Record')
plt.xticks(rotation=0)
plt.show()
```

# 3.What is the distribution of assets, liabilities, and net worth of the elected MPs?

```python
# Remove non-numeric characters from 'ASSETS' column
df['ASSETS'] = df['ASSETS'].str.replace(r'[^\d]+', '', regex=True)
```

```python
# Convert 'ASSETS' to integers
df['ASSETS'] = pd.to_numeric(df['ASSETS'], errors='coerce')
```

```python
df["ASSETS"].unique()
```

```python
# Remove non-numeric characters from 'LIABILITIES' column
df["LIABILITIES"]=df["LIABILITIES"].str.replace(r'[^\d]+',"",regex=True)
```

```python
# Convert 'LIABILITIES' to integers
df['LIABILITIES'] = pd.to_numeric(df['LIABILITIES'], errors='coerce')
```

```python
df["LIABILITIES"].unique()
```

```python
# Calculate net worth (assets - liabilities)
df['NET WORTH'] = df['ASSETS'] - df['LIABILITIES']
```

```python
# Create subplots for each column
fig, axes = plt.subplots(1,3,figsize=(12, 4))

# Plot histogram for NET WORTH
axes[0].hist(df['NET WORTH'].dropna(), bins=30, color='blue', alpha=0.7)
axes[0].set_title('Histogram of NET WORTH for Elected MPs')
axes[0].set_xlabel('NET WORTH')
axes[0].set_ylabel('Frequency')

# Plot histogram for ASSETS
axes[1].hist(df['ASSETS'].dropna(), bins=30, color='green', alpha=0.7)
axes[1].set_title('Histogram of ASSETS for Elected MPs')
axes[1].set_xlabel('ASSETS')
axes[1].set_ylabel('Frequency')

# Plot histogram for LIABILITIES
axes[2].hist(df['LIABILITIES'].dropna(), bins=30, color='red', alpha=0.7)
axes[2].set_title('Histogram of LIABILITIES for Elected MPs')
axes[2].set_xlabel('LIABILITIES')
axes[2].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()
plt.show()
```

# 5.What is the region-wise distribution of seats across the country?

```python
region_distribution=df['REGION'].value_counts()
```

```python
# Plot the bar graph
plt.figure(figsize=(12, 6))
sns.barplot(x=region_distribution.index, y=region_distribution.values, palette='viridis')
plt.title('Region-wise Distribution of Seats Across States')
plt.xlabel('Region')
plt.ylabel('Number of Seats')
plt.xticks(rotation=0)  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()
```

```python
region_winning=df.groupby('REGION')['WINNER'].sum()
```

```python
# Create distribution plot for region-wise winners
plt.figure(figsize=(12, 6))
plt.bar(region_winning.index,region_winning.values, color='lightgreen')
plt.title('Region-wise Election Winners')
plt.xlabel('Regions')
plt.ylabel('Winning Candidates')
plt.show()
```
